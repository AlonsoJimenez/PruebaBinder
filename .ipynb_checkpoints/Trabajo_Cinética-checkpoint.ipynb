```python
from typing import Container
from IPython.display import HTML
from IPython.display import display

from ipywidgets import interactive
import ipywidgets as widgets
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches


# Tomado de https://stackoverflow.com/questions/31517194/how-to-hide-one-specific-cell-input-or-output-in-ipython-notebook
tag = HTML('''<script>
code_show=false; 
function code_toggle() {
    if (code_show){
        $('div.cell.code_cell.rendered.selected div.input').hide();
    } else {
        $('div.cell.code_cell.rendered.selected div.input').show();
    }
    code_show = !code_show
} 

$( document ).ready(code_toggle);
</script>

Para mostrar/ocultar código presione <a href="javascript:code_toggle()">aquí</a>.''')


def Trabajo():

    display(tag)
    
    def GraficoInteractivoTrabajo(angulo, fuerza, tiempo):

        anguloRadianes = angulo * np.pi / 180
        desplazamiento = tiempo 

        
        trabajo = desplazamiento * fuerza * np.cos(anguloRadianes)
    
        fig, ax1 = plt.subplots(dpi=120)   
        ax1.set_xlim ([-2, 19])
        ax1.set_ylim ([-1.5, 3.5])
        ax1.set_title('Trabajo en una partícula')


        # Caja
        ax1.add_patch(
            patches.Rectangle((-1.5 + tiempo, -0.25), 2, 1,
                facecolor = 'red',fill=True) )

        # Cuerda
        x = [tiempo + 0.5, tiempo + 0.5 + np.cos(anguloRadianes)]
        y = [0.25, 0.25 + np.sin(anguloRadianes)]

        ax1.plot (x, y)
        
        #Flecha y texto para la fuerza.
        ax1.arrow(x[1], y[1], fuerza/5 * np.cos(anguloRadianes), fuerza/5 *  np.sin(anguloRadianes), head_width=0.35, head_length=0.1,
                  fc='k', ec='k')
        ax1.text(x[1]+(fuerza/5+0.25) * np.cos(anguloRadianes), y[1]+0.25+(fuerza/5+0.25) *  np.sin(anguloRadianes), 
                 "Fuerza=" + str(fuerza) + "N")

        #Flecha y texto para el desplazamiento.
        ax1.arrow(0.5, -0.5, tiempo, 0, head_width=0.35, head_length=0, fc='k', ec='k')
        ax1.arrow(0.5+tiempo, -0.5, -tiempo, 0, head_width=0.35, head_length=0, fc='k', ec='k')
        ax1.text((x[1])/2-1, -1.25, "Desplazamiento=" + str(round (desplazamiento,2)) + "m")


        # Barra Trabajo 
        ax1.add_patch(
            patches.Rectangle((17, 1), 2,  trabajo/25,
                facecolor = 'c',fill=True) )
        #Resultado Trabajo
        ax1.text(16, 3.5, "Trabajo=" + str(round (trabajo,1)) + "J")
        ax1.text(19.25, 1, 0)
        
        
        #Resultado Trabajo
        ax1.text(-1, 2.75, "Trabajo=" + str(round (trabajo,2)) + "J")

        ax1.axis ("off")

        plt.show()  
        return

    angEtiqueta = widgets.Label (value="Angulo $(°)$:")
    angSlider = widgets.FloatSlider(min=0, max=180, step=1, value=45)
    cajaAngulo = widgets.HBox([angEtiqueta, angSlider])
    
    fuerzaEtiqueta = widgets.Label (value="Fuerza $(N)$:")
    fuerzaSlider = widgets.FloatSlider(min=1, max=5, step=0.5, value=3)
    cajaFuerza = widgets.HBox([fuerzaEtiqueta, fuerzaSlider])

    play = widgets.Play(value=0, min=0, max=30, step=1, disabled=False)
    tieEtiqueta = widgets.Label (value="Tiempo $(s)$:")
    tieSlider = widgets.FloatSlider(min=0, max=10,)
    widgets.jslink((play, 'value'), (tieSlider, 'value'))
    cajaTiempo = widgets.HBox([tieEtiqueta, tieSlider])


    salida = widgets.interactive_output(GraficoInteractivoTrabajo, {'angulo':angSlider,
                                                                     'fuerza':fuerzaSlider,
                                                                     'tiempo':tieSlider,})

    display(cajaAngulo, cajaFuerza, cajaTiempo, play, salida)


def Resorte():

    display(tag)
    
    def GraficoInteractivoResorte(desplazamiento, constante):

        fuerza = -constante * desplazamiento

        trabajo = fuerza * desplazamiento / 2 

        fig, ax = plt.subplots (dpi=120)    
        plt.xlim (-6, 23)
        plt.ylim (-3, 2)
        ax.set_title('Trabajo en un resorte')

        # Pared
        ax.add_patch(
            patches.Rectangle((-6, -1), 6, 2,
                facecolor = 'gray',fill=True) )

        deltaX = desplazamiento + 7
        x = np.linspace (0, deltaX,100)
        y = np.sin(20*np.pi*x/deltaX)  

        #Flecha y texto para la fuerza.
        ax.arrow(7, 1.25, fuerza/25, 0, head_width=0.35, head_length=0.3, fc='k', ec='k')
        ax.text(4, 1.5, "Fuerza=" + str(round(fuerza,2)) + "N")

        #Flecha y texto para el desplazamiento.
        ax.arrow(7, -1.25, desplazamiento, 0, head_width=0.35, head_length=0, fc='k', ec='k')
        ax.arrow(7 + desplazamiento, -1.25, -desplazamiento, 0, head_width=0.35, head_length=0, fc='k', ec='k')
        ax.text(0, -1.75, "Desplazamiento=" + str(round(desplazamiento,2)) + "m")

        # Barra Trabajo 
        ax.add_patch(
            patches.Rectangle((18, 1.25), 4,  trabajo/100,
                facecolor = 'c',fill=True) )
        #Resultado Trabajo
        ax.text(16.5, 1.5, "Trabajo=" + str(round (trabajo,1)) + "J")

        ax.plot (x,y)  

        ax.axis ("off")

        plt.show()  
        return

    despEtiqueta = widgets.Label (value="Desplazamiento $(m)$:")
    despSlider = widgets.FloatSlider(min=-5, max=5, step=0.1, value=0)
    cajaDesp = widgets.HBox([despEtiqueta, despSlider])
    
    constanteEtiqueta = widgets.Label (value="Constante elástica $(N/m)$:")
    constanteSlider = widgets.FloatSlider(min=10, max=30, step=1, value=20)
    cajaConstante = widgets.HBox([constanteEtiqueta, constanteSlider])

    salida = widgets.interactive_output(GraficoInteractivoResorte, {'desplazamiento':despSlider,
                                                                     'constante':constanteSlider})

    display(cajaDesp, cajaConstante, salida)


def Caída():

    display(tag)
    
    def GraficoInteractivoCinético(masa, tiempo):

        tiempo = tiempo / 1000
        altura = 19.6

        posición = -9.8/2 * tiempo**2

        velocidad = -9.8 * tiempo

        energía = 1/2 * masa * velocidad**2

        fig, ax = plt.subplots (dpi=120)    
        plt.xlim (0, 5)
        plt.ylim (-7, 35)
        ax.set_title('Energía cinética en una partícula')

        # Caja
        ax.add_patch(
            patches.Rectangle((1, altura + posición), 0.15, 2,
                facecolor = 'red',fill=True) )


        #Flecha y texto para la velocidad.
        ax.arrow(1.2, altura + posición, 0, velocidad / 3, head_width=0.05, head_length=0.3, fc='k', ec='k')
        ax.text(1.25, altura + posición + velocidad / 3, "Velocidad=" + str(round(-velocidad,2)) + "m/s")

        # Barra
        ax.add_patch(
            patches.Rectangle((3, 0), 1, energía/175,
                facecolor = 'c',fill=True) )

        ax.text(4.1, energía/175, "Energía=" + str(round(energía,2)) + "J")

 
        ax.axis ("off")

        plt.show()  
        return
    
    masaEtiqueta = widgets.Label (value="Masa $(kg)$:")
    masaSlider = widgets.FloatSlider(min=10, max=30, step=1, value=20)
    cajaMasa = widgets.HBox([masaEtiqueta, masaSlider])

    play = widgets.Play(value=0, min=0, max=2000, step=100, disabled=False)
    tieEtiqueta = widgets.Label (value="Tiempo $(ms)$:")
    tieSlider = widgets.FloatSlider(min=0, max=2000,)
    widgets.jslink((play, 'value'), (tieSlider, 'value'))
    cajaTiempo = widgets.HBox([tieEtiqueta, tieSlider])


    salida = widgets.interactive_output(GraficoInteractivoCinético, {'masa':masaSlider,
                                                                     'tiempo':tieSlider,})

    display(cajaMasa, cajaTiempo, play, salida)


def Sistema():

    display(tag)
    
    def GraficoInteractivoSistema(constante, masa, tiempo):
        tiempo = tiempo / 1000

        w = np.sqrt (constante / masa)

        desplazamiento = 5 * np.cos (w * tiempo) 

        fuerza = -constante * desplazamiento

        trabajo = fuerza * desplazamiento / 2

        cinética = -constante * 25 / 2 - trabajo

        fig, ax = plt.subplots (dpi=120)    
        plt.xlim (-6, 32)
        plt.ylim (-2.25, 2)
        ax.set_title('Energías en una caja atada a un resorte')

        # Pared
        ax.add_patch(
            patches.Rectangle((-6, -1), 6, 2,
                facecolor = 'gray',fill=True) )
        # Resorte
        deltaX = desplazamiento + 7
        x = np.linspace (0, deltaX,100)
        y = np.sin(20*np.pi*x/deltaX)  
        
        # Caja
        ax.add_patch(
            patches.Rectangle((deltaX, -0.15), 2, 0.3,
                facecolor = 'red',fill=True) )


        # Barra Trabajo 
        ax.add_patch(
            patches.Rectangle((18, -2), 4, - trabajo/75,
                facecolor = 'c',fill=True) )
        #Resultado Trabajo
        ax.text(16.5, 1.5, "Trabajo=" + str(round (-trabajo,1)) + "J")

        # Barra Cinética 
        ax.add_patch(
            patches.Rectangle((27.5, -2), 4, - cinética/75,
                facecolor = 'green',fill=True) )
        #Resultado Cinética
        ax.text(26, 1.5, "Energía Cinética=" + str(round (-cinética,1)) + "J")

        ax.plot (x,y)  

        ax.axis ("off")

        plt.show()  
        return
    
    constanteEtiqueta = widgets.Label (value="Constante elástica $(N/m)$:")
    constanteSlider = widgets.FloatSlider(min=10, max=30, step=1, value=20)
    cajaConstante = widgets.HBox([constanteEtiqueta, constanteSlider])

    masaEtiqueta = widgets.Label (value="Masa $(kg)$:")
    masaSlider = widgets.FloatSlider(min=10, max=30, step=0.1, value=0)
    cajaMasa = widgets.HBox([masaEtiqueta, masaSlider])

    play = widgets.Play(value=0, min=0, max=20000, step=100, disabled=False)
    tieEtiqueta = widgets.Label (value="Tiempo $(ms)$:")
    tieSlider = widgets.FloatSlider(min=0, max=20000,)
    widgets.jslink((play, 'value'), (tieSlider, 'value'))
    cajaTiempo = widgets.HBox([tieEtiqueta, tieSlider])

    salida = widgets.interactive_output(GraficoInteractivoSistema, {'constante': constanteSlider,
                                                                     'masa':masaSlider,
                                                                     'tiempo':tieSlider,})

    display(cajaConstante, cajaMasa, cajaTiempo, play, salida)
```

# Trabajo

Se puede definir, que cuando una fuerza efectúa un trabajo es debido a un desplazamiento del centro de masa del cuerpo en el cual se está efectuando la fuerza.



Este matemáticamente se define como:
$$
dW=\overrightarrow{F} \cdot d\overrightarrow{x}
$$

Si se integra a ambos lados de la ecuación se obtiene que:

$$
\int_{}^{} \, dW = \displaystyle\int_{x_1}^{x_2} \overrightarrow{F}\, d\overrightarrow{x}
$$

Por lo que:

$$
W = \overrightarrow{F}\Delta x
$$

Es importante recalcar que el trabajo es un escalar definido por un producto punto, por lo que también se puede definir como:

$$
dW= F\Delta xcos(\theta)
$$



Como se muestra en la siguiente imagen:



Si la fuerza tiene dirección opuesta al desplazamiento, el trabajo será negativo.

# Ley de Hooke

Es la fuerza ejercida por un resorte que es comprimido o estirado, la cual se puede representar como:

$$
\overrightarrow{F} = -k\Delta \overrightarrow{x}
$$

Donde $k$ es la constante elástica del resorte. 



Con base a lo anterior, se puede definir el trabajo realizado por un resorte como:

$$
\int_{}^{} \, dW = \displaystyle\int_{x_1}^{x_2} -k\overrightarrow{x} \cdot \, d\overrightarrow{x}
$$

Es importante recalcar que la fuerza de restitución ejercida por el resorte debido a la Ley de Hooke, es antiparalela al dezplazamiento del mismo, por lo cual el producto punto posee un resultado negativo. Luego el trabajo es igual a:

$$
W = -\dfrac{1}{2}k (\Delta x)^2
$$

# Energía Cinética

La energía cinética se define como el trabajo necesario para acelerar una masa. En otras palabras la energía asociada al movimiento de un cuerpo. La cual se puede escribir como:

$$
K = \dfrac{1}{2}mv^2
$$
